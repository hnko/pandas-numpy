```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'HPI':[80, 85, 88, 85],
                   'Int_rate': [2, 3, 2, 2],
                   'Us_GDP_Thousands': [50, 55, 65, 55]},
                  index = [2001, 2002, 2003, 2004])

df2 = pd.DataFrame({'HPI':[80, 85, 88, 85],
                   'Int_rate': [2, 3, 2, 2],
                   'Us_GDP_Thousands': [50, 55, 65, 55]},
                  index = [2005, 2006, 2007, 2008])

df3 = pd.DataFrame({'HPI':[80, 85, 88, 85],
                   'Int_rate': [2, 3, 2, 2],
                   'Low_tier_HPI': [50, 52, 50, 53]},
                  index = [2001, 2002, 2003, 2004])
```

```python
#concatenation - df1 and df2 - look the index
concat = pd.concat([df1, df2])
concat
```

```python
concat2 = pd.concat([concat, df3])
concat2
```

```python
df4 = df1.append(df2)
df4
```

```python
df1.append([df2, df3])
```
