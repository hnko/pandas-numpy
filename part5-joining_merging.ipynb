```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2,3,2,2],
                    'US_GDP_Thousands':[50,55,65,55]},
                     index=[2001,2002,2003,2004])
df2 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2,3,2,2],
                    'US_GDP_Thousands':[50,55,65,55]},
                     index=[2005,2006,2007,2008])
df3 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Unemployment':[7,8,9,6],
                    'Low_tier_HPI':[50,52,50,53]},
                     index=[2001,2002,2003,2004])
```

```python
#merge
pd.merge(df1,df2, on='HPI')
```

```python
pd.merge(df1,df2, on=['HPI', 'Int_rate'])
```

```python
df1.set_index('HPI', inplace=True)
df3.set_index('HPI', inplace=True)
```

```python
joined = df1.join(df3)
joined
```

```python
df1 = pd.DataFrame({'year':[2001,2002,2003,2004],
                    'Int_rate':[2,3,2,2],
                    'US_GDP_Thousands':[50,55,65,55]})

df3 = pd.DataFrame({'year':[2001,2003,2004,2005],
                    'Unemployment':[7,8,9,6],
                    'Low_tier_HPI':[50,52,50,53]})
```

```python
merged = pd.merge(df1, df3, on='year')
merged.set_index('year', inplace=True)
merged
```

```python
pd.merge(df1, df3, on='year', how='left')
```

```python
pd.merge(df1, df3, on='year', how='right')
```

```python
pd.merge(df1, df3, on='year', how='outer')
```

```python
pd.merge(df1, df3, on='year', how='inner')
```

# Joining

```python
#usefull to preserve the order when we create de df
from collections import OrderedDict

employees = pd.DataFrame(OrderedDict({'id':[2,6,7,12],
                        'name':['chris', 'sandy','christine','aaron'],
                         'title_id':[1,2,1,3],
                         'years':[15,5,5,3],
                         'department_id':[2,7,4,3]}))
```

```python
employees.set_index('id', inplace=True)
```

```python
employees
```

```python
titles = pd.DataFrame(OrderedDict({'id':[1,2,3],
                                  'name':['President', 'Lead Engineer', 'Intern']}))
```

```python
titles.set_index('id', inplace=True)
```

```python
titles
```

```python
employees.join(titles, rsuffix='_title')
```

```python
employees
```

```python
employees.join(titles, on=['title_id'], rsuffix='_title')
```

```python
departments = pd.DataFrame(OrderedDict({'id':[1,2,3,4],
                                       'name':['sales','product','technical support', 'human resources']}))
departments.set_index('id', inplace=True)
departments
```

```python
employees.join(titles, on=['title_id'], rsuffix='_title').join(departments, on=['department_id'], rsuffix='_departments')
```

```python
employees.join(titles, on=['title_id'], rsuffix='_title').join(departments, on=['department_id'], how='inner', rsuffix='_departments')
```
