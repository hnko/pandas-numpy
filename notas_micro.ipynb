```python
import pandas as pd
```

```python
df = pd.read_csv('notasMicro2018.csv', sep=';', encoding='ISO-8859-1' ,names=['1', 'P1', 'P2', '2', '3', 'P3', 'Teoria', 'Practica', 'Definitivo'])
```

```python
#delete the columns we are not going to use
df.drop(df.columns[[0, 3, 4]], axis=1, inplace=True)
df.drop(df.index[0], inplace=True)
```

```python
df.head()
```

```python
# convert the values to numerib values
for i in range(6):
    df[df.columns[i]] = pd.to_numeric(df[df.columns[i]].str.replace(',', '.'), errors='coerce') # invalid parsing will be set as NaN
```
